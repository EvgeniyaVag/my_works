```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('/content/09.11.23+book_simple.db (1).db')
cur = con.cursor()
```

```python
###
# 1.Выведите авторов, имеющихся в базе.

req1 = """
SELECT name
FROM author;
"""

pd.read_sql(req1, con)
```

```python
###
# 2. Найдите всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия

req2 = """
SELECT name
FROM author
WHERE name LIKE 'С. А._%';
"""

pd.read_sql(req2, con)
```

```python
###
# 3. Выведите список книг, выпущенных в издательстве "Мир".

req3 = """
SELECT book.title,
       publisher.name
FROM book, publisher
WHERE (book.publisher_id = publisher.id) AND
       (publisher.name = 'Мир');
"""

pd.read_sql(req3, con)
```

```python
###
# 4. Выведите количество книг, принадлежащих каждой из категорий. Результат отсортировать по возрастанию

req4 = """
SELECT COUNT(book.id) AS Количество_книг,
       category.title AS Категория
FROM book, category
WHERE (book.category_id = category.id)
GROUP BY category.title
ORDER BY COUNT(book.id) ASC;
"""

pd.read_sql(req4, con)
```

```python
###
# 5. Выведите количество книг, выпущенных в каждом издательстве.
#В итоговом списке оставьте только те издательства,
#которые выпустили 3 и более книг. Результат отсортировать по убыванию.

req5 = """
SELECT COUNT(book.id) AS 'Количество книг',
       publisher.name AS Издательство
FROM book, publisher
WHERE (book.publisher_id = publisher.id)
GROUP BY publisher.name
HAVING COUNT(book.id) >= 3
ORDER BY COUNT(book.id) DESC;
"""

pd.read_sql(req5, con)
```

```python
### Вариант_1
# 6. Выведите список книг, автором которых являлся <<С. А. Айвазян>>

req6 = """
SELECT book.title AS 'Название книги',
       author.name AS 'Автор книги'
FROM book
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE author.name LIKE '%Айвазян%';
"""

pd.read_sql(req6, con)
```

```python
### ВАРИАНТ_2
# 6. Выведите список книг, автором которых являлся <<С. А. Айвазян>>

req6_2 = """
SELECT book.title AS 'Название книги',
       author.name AS 'Автор книги'
FROM book, author, authorbook
WHERE (authorbook.book_id = book.id)
      AND (authorbook.author_id = author.id)
      AND (author.name LIKE '%Айвазян%');
"""

pd.read_sql(req6_2, con)
```

```python
###
# 7. В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>.
#Результат отсортировать по убыванию

req7 = """
SELECT publisher.name AS 'Название издательства',
       COUNT(book.id) AS 'Количество книг',
       author.name AS Автор
FROM book
INNER JOIN publisher
ON book.publisher_id = publisher.id
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE author.name LIKE '%Айвазян%'
GROUP BY publisher.name
ORDER BY COUNT(book.id) DESC;
"""

pd.read_sql(req7, con)
```

```python
### Вариант_2
# 7. В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>.
#Результат отсортировать по убыванию

req7_2 = """
SELECT publisher.name AS 'Название издательства',
       COUNT(book.id) AS 'Количество книг',
       author.name AS Автор
FROM book, publisher, author, authorbook
WHERE (authorbook.book_id = book.id)
      AND (authorbook.author_id = author.id)
      AND (book.publisher_id = publisher.id)
      AND (author.name LIKE '%Айвазян%')
GROUP BY publisher.name
ORDER BY COUNT(book.id) DESC;
"""

pd.read_sql(req7_2, con)
```
