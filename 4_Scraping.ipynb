```python
soup = BeautifulSoup(answer.text, 'lxml')
print(soup.prettify())
```

```python
prod_card = [] #Список всех продуктовых карточек

prod_card = soup.find_all('div', attrs = {'class': ['product-card js-item-card']}) #class_ = 'product-card js-item-card')
#print(len(prod_card))
#print(prod_card)
```

```python
print(prod_card[7].find('a', attrs = {'class':'ttl'}).get_text(strip=True)) #Взяли название товара
print(prod_card[7].find('span', attrs = {'class':['price hot-price','price','val',None]}).get_text(strip=True)) #Взяли цену
prod_card[7].find('a').get('href') #Взяли ссылку
```

```python
dict_df = {
           'Наименование товара': [],
           'Ссылка' : [],
           'Цена (руб.)' : []
          }

for product in prod_card:
    dict_df['Наименование товара'].append(product.find('a', attrs = {'class':'ttl'}).get_text(strip=True)) #Взяли название товара
    dict_df['Ссылка'].append(product.find('a').get('href')) #Взяли ссылку
#    dict_df['price'].append(product.find('span', attrs = {'class':'price hot-price'}).get_text(strip=True)) #Взяли цену
    dict_df['Цена (руб.)'].append(product.find('span', attrs = {'class':['price hot-price','price','val','last-price']}).get_text(strip=True)) #Взяли цену
```
